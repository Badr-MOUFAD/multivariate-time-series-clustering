```python
# imports
# scientific / data manipulation
import numpy as np
import pandas as pd

# plots
from plotly import graph_objects as go
import plotly.express as px
import plotly.io as pio

pio.templates.default = "plotly_white"
```

```python
# load data
# skip rows in the header (0 - 15) --> range(16)

li_rows_skip = list(range(16))
data = pd.read_csv("/drive/My Drive/Colab Notebooks/AgriEdge/datasets/weather_data_marrakech.csv", skiprows=li_rows_skip)
```

```python
# plot temperature

# build query
arr_years = [2000, 2015]
query = data["YEAR"].isin(arr_years)

# select and plot
data_year = data[query]
fig = px.line(data_year, x="DOY", y="T2M_MAX", color="YEAR")

fig.show()
```

```python
# evolution of temprature
col_name = "PRECTOT"

fig = px.line(
    data, x="DOY", y=col_name, 
    animation_frame="YEAR", 
    range_y=[data[col_name].min() - 10, data[col_name].max() + 10],)

fig.show()
```

```python
import numpy as np

center_point = np.array([34.269355, -5.938411], dtype=float)
perimeter_point = np.array([34.26860, -5.99298], dtype=float)

r = np.sqrt((center_point - perimeter_point) @ (center_point - perimeter_point))
```

```python
def gen_point(theta):
  return center_point + r * np.array([np.cos(theta), np.sin(theta)])


gen_point(np.pi)
```
