```python
# read data from dir
# construct dict

import os

def construct_weather_arr(dir_path="/drive/My Drive/Colab Notebooks/AgriEdge/weather_dataset/"):
  dict_locations_weather = []

  for filename in os.listdir(dir_path):
    dict_locations_weather.append(pd.read_csv(dir_path + filename))
  
  return dict_locations_weather
```

```python
locations_weather = construct_weather_arr()
```

```python
# test whether locations have the same recordings

def col_std(arr_data_frame, col_name, ):
  # compute a mean col
  mean_col = arr_data_frame[0][col_name]

  n = len(arr_data_frame)
  for i in range(1, n):
    mean_col = mean_col + arr_data_frame[i][col_name]

  mean_col = mean_col / n

  # compute a std
  std_col = (arr_data_frame[0][col_name] - mean_col) ** 2

  for i in range(1, n):
    std_col = std_col + (arr_data_frame[i][col_name] - mean_col) ** 2

  std_col = np.sqrt(std_col / n)

  # return max of col
  return max(std_col)
```

```python
col_std(arr_data_frame=locations_weather, col_name="T2M_MIN")
```

```python
ARR_COLS = ["ALLSKY_SFC_SW_DWN", "PRECTOT", "RH2M", "T2M", "T2MDEW", "T2M_MAX", "T2M_MIN", "WS2M"]

for col in ARR_COLS:
  print(col_std(arr_data_frame=locations_weather, col_name=col))
```
